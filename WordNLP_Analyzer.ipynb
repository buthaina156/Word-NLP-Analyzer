```python
print("Hello, I'm word teacher.ğŸ‘©â€ğŸ«ğŸ‘‹")
print("Type one word and I will analyze it with NLP")
print("\n********************************************")

while True:
  user_word = input("\n Type the word you want to analyze or enter 'quit' if you finished:").lower()
  if user_word == 'quit':
    print(" Bye bye ğŸ‘‹ğŸ¼")
    break
  w = Word(user_word)

  # Check spelling
  corrected = w.spellcheck()[0][0]

  # Lemmatization (base from)
  lemma = Word(corrected).lemmatize('v')


  print(f"\nğŸ” NLP Analysis for '{user_word}':")
  print(f"âœ… Corrected spelling: {corrected}")
  print(f"ğŸŒ± Lemma (base form): {lemma}")
```
